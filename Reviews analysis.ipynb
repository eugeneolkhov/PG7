```python
#Hey this is eugene, just testing to see if it works

import pandas as pd
import numpy as np

df = pd.read_csv('OnlineNewsPopularity.csv')
df.head()
```

```python
## Dropping URL variable & using describe###
df.drop('url', axis = 1, inplace = True)
df = df.astype(int)
df.describe().T
```

```python
## checking number of unique values for each var ###
for i in df:
    print(i,(df[i].nunique()))
```

```python
## making token titles object for prediction ####
df[' n_tokens_title'] = df[' n_tokens_title'].astype(object)
```

```python
## plots on plots ####

import matplotlib.pyplot as plt
import seaborn as sns

#df["n_tokens_title"] = df["n_tokens_title"].astype(object)
sns.barplot(x = " n_tokens_title", y = " shares", data = df)
plt.show()
```

```python
sns.barplot(x = " title_sentiment_polarity", y = " shares", data = df)
plt.show()
```

```python
import seaborn as sns
a = df[(df[' shares'] <= 25000) & (df[" n_tokens_content"] <= 4000)]
sns.regplot(x = " n_tokens_content", y = ' shares', data = a, scatter_kws={'alpha':0.2})
plt.show()
```

```python
import seaborn as sns
a = df.loc[(df[' shares'] <= 6500) & (df[" n_tokens_content"] <= 4000), [" n_tokens_title", ' shares']]
sns.boxplot(x = " n_tokens_title", y = ' shares', data = a)
plt.show()
```
